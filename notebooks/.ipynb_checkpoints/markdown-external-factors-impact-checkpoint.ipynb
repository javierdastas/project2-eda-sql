```python
# Import necessary libraries
import pymysql
import pandas as pd
import matplotlib.pyplot as plt
```

```python
database_name = "retail_data" # Database name

# Connet to DataBase
cnx = pymysql.connect(user='root', password='root', host='localhost', database=database_name) # Authentication
if cnx.open:
    print("Connection open")
else:
    print("Connection is not successfully open")
```

# Data Analysis and Visualization with SQL, Pandas and Pyplot

## Question 1 - Markdown Impact: Do weeks with higher MarkDown values have higher average sales?

```python
# SQL Query to get MarkDown1 and Weekly_Sales 
query = """
SELECT 
    f.MarkDown1,
    f.MarkDown2,
    f.MarkDown3,
    f.MarkDown4,
    f.MarkDown5,
    s.Weekly_Sales
FROM 
    features f
JOIN 
    sales s ON f.Store = s.Store AND f.Date = s.Date
WHERE 
    f.Date >= '2011/11/01';
"""
# Query execute 
cursor = cnx.cursor()
cursor.execute(query)
data_query = cursor.fetchall()
```

```python
# Import from the Table to the DataFrame
columns = [desc[0] for desc in cursor.description]
markdown_weekly_sales_df = pd.DataFrame(data_query, columns=columns)
```

```python
# Close Database Conection
if cnx.open: # Check if connection is open
    cursor.close()
    cnx.close()
```

```python
# Set the ranges and labels 
combined_values = pd.concat([
    markdown_weekly_sales_df[f'MarkDown{i}'] for i in range(1, 6) # Use adjusted percentil for the ranges
])
bins = pd.qcut(combined_values, q=4, duplicates='drop', retbins=True)[1]
labels = ['Low', 'Medium', 'High', 'Very High']

# set the columns range per markdown
for i in range(1, 6):
    column_name = f'MarkDown{i}'
    bin_column = f'{column_name}_Bin'
    markdown_weekly_sales_df[bin_column] = pd.cut(
        markdown_weekly_sales_df[column_name], bins=bins, labels=labels
    )

# Get the Weekly_Sale Average for each range (bin) and MarkDown
average_sales_dict = {}
for i in range(1, 6):
    bin_column = f'MarkDown{i}_Bin'
    average_sales = markdown_weekly_sales_df.groupby(bin_column, observed=True)['Weekly_Sales'].mean()
    average_sales_dict[f'MarkDown{i}'] = average_sales

# Merge the Data inside a DataFrame
average_sales_df = pd.DataFrame(average_sales_dict).reset_index()

# Create the  graph
x = average_sales_df['index']  # Rango de MarkDown (Low, Medium, High, Very High)
width = 0.15  # Ancho de cada barra
positions = range(len(x))

plt.figure(figsize=(10, 6))

# Create a bar by MarkDown
for i, column in enumerate(average_sales_dict.keys()):
    plt.bar(
        [p + i * width for p in positions],  # Desplazar cada barra según su grupo
        average_sales_df[column],  # Altura de las barras
        width=width,
        label=column
    )

# Config Graph Details
print()
plt.title('MarkDown Weekly Sales Average Comparison by Rank', fontsize=16)
plt.xlabel('MarkDown Ranges', fontsize=14)
plt.ylabel('Average Weekly Sales', fontsize=14)
plt.xticks([p + width * 2 for p in positions], x, fontsize=12)  # Centrar las etiquetas
plt.legend(title='Lengend', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

## Question 1 - Markdown Impact: Do weeks with higher MarkDown values have higher average sales?

## Answer based on the provided chart:

- MarkDown1: Yes, there is a clear positive trend where weeks with higher MarkDown1 values show higher average sales. The bars increase steadily as the range moves from "Low" to "Very High."
- MarkDown2: There is no consistent pattern. While there is an increase in the "Low" category, the trend does not clearly indicate higher average sales for higher MarkDown2 values.
- MarkDown3: The "Medium" category stands out with a sharp increase, but the trend is not consistent across other ranges, and the relationship is unclear.
- MarkDown4: Shows mixed behavior with slight increases in average sales for higher ranges, but the pattern lacks consistency.
- MarkDown5: Displays a positive trend, especially in the "Very High" category, where average sales peak significantly.

## Conclusion: 
Not all MarkDown variables exhibit a consistent positive impact on average sales. MarkDown1 and MarkDown5 show a strong correlation with higher average sales as their values increase, whereas MarkDown2, MarkDown3, and MarkDown4 do not demonstrate a clear or consistent pattern.
